```python

space = "\n\t\t\t\t\t"
separate = "\n--------------------------------------------------------------------------------------------------------"
options = ('1', '2', '3', '4', '5', '6' ,'exit') 


def Menu():
        print("Select operation.\n")
        print("1.Add")
        print("2.Subtract")
        print("3.Multiply")
        print("4.Divide")
        print("5.Power")
        print("6.Square")
        
        
class Calculator:
    
    def __init__(self,Number1,Number2):
        self.Number1 = Number1
        self.Number2 = Number2
    
    
    def add(self,x,y):
        
        return x + y 
    
    def subtract(self,x,y):
        
        return x - y
    
    def multiply(self,x,y):
        
        return x * y
    
    def power (self,x,y):
        
        return x ** y
    
    def square (self,x,y):
        
        return self.power (x,2)
    
    def divide(self,x,y): 
        if y == 0: 
            return "You can't divide by zero!" 
        else:
            return x/y 
        
                
    def Result(self,choice):
            if    choice ==    '1' : return  str(x) + " + " + str(y) + " = " + str(calcul.add(x,y))
            elif  choice ==    '2' : return  str(x)+ " - "+ str(y)+ " = "+str(calcul.subtract(x,y))
            elif  choice ==    '3' : return  str(x)+ " * "+ str(y)+ " = "+str(calcul.multiply(x,y))
            elif  choice ==    '4' : return  str(x)+ " / "+ str(y)+ " = "+str(calcul.divide(x, y))
            elif  choice ==    '5' : return  str(x)+ " ** "+ str(y)+ " = "+str(calcul.power(x, y))
            elif  choice ==    '6' : return  str(x)+ " ^ "+ str(x)+ " = "+str(calcul.square(x, y))
                
    def Border(self,string, length):
            return (string[0+i:length+i] for i in range(0, len(string), length))
        
    
        
        
while True:
    
        Menu()
       
        choice = input("\nEnter your choice (1/2/3/4/5/6/exit) : ")

        if choice in options[0:-1]:

        # Numbers input 

            try :
                x = float(input("\nEnter the first number: "))

                if choice != options[-2]: 
                    y = float(input("Enter the second number: "))

            except ValueError:
                print("Please enter a valid number !!!")
                print (separate)
                continue

            calcul = Calculator(x,y)
           
        # Diplay Format
        
            width = len(calcul.Result(choice))

            print(space,'+-' + '-' * width + '-+')

            for line in calcul.Border(calcul.Result(choice), width):
                print(space,'| {0:^{1}} |'.format(line, width))

            print(space,'+-' + '-'*(width) + '-+')

        # Exit
        
        elif choice == options[-1]:
            break;
            
        else:
            print('\n choose a valid option !! \n')
            
        print (separate)
    

        

```
