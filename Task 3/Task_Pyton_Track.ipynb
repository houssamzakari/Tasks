```python
import math

space = "\n\t\t\t\t\t"
separate = "\n--------------------------------------------------------------------------------------------------------"
options = ('0','1', '2', '3', '4', '5', '6' ,'exit') 


Menu = """Select an operation.\n
        0.All Operations 
        1.Add
        2.Subtract
        3.Multiply
        4.Divide
        5.Power
        6.Square root
        exit"""

def Display (calcul,count):
    
    while True:
            
            Result = calcul.Result(count)
            
            width = len(Result)

            print(space,'+-' + '-' * width + '-+')

            for line in calcul.Border(Result, width):
                print(space,'| {0:^{1}} |'.format(line, width))

            print(space,'+-' + '-'*(width) + '-+')

            if choice == '0' and int(count) < 6 :
                count = str(int(count)+1)
            else:
                break;
                
                                              
        
class Calculator:
    
    def __init__(self,Number1,Number2):
        self.Number1 = Number1
        self.Number2 = Number2
    
    
    def add(self,x,y):
        
        return x + y 
    
    def subtract(self,x,y):
        
        return x - y
    
    def multiply(self,x,y):
        
        return x * y
    
    def power (self,x,y):
        
        return x ** y
    
    def square_root (self,x,y):
        
        return math.sqrt(x)
    
    def divide(self,x,y): 
        if y == 0: 
            return "You can't divide by zero!" 
        else:
            return x/y 
        
                
    def Result(self,choice):
            if  choice ==    '1' : return  str(x) + " + " + str(y) + " = " + str(calcul.add(x,y))
            if  choice ==    '2' : return  str(x)+ " - "+ str(y)+ " = "+str(calcul.subtract(x,y))
            if  choice ==    '3' : return  str(x)+ " * "+ str(y)+ " = "+str(calcul.multiply(x,y))
            if  choice ==    '4' : return  str(x)+ " / "+ str(y)+ " = "+str(calcul.divide(x, y))
            if  choice ==    '5' : return  str(x)+ " ** "+ str(y)+ " = "+str(calcul.power(x, y))
            if  choice ==    '6' : return  " √ "+ str(x)+ " = "+str(calcul.square_root(x, y))

    def Border(self,string, length):
            return (string[0+i:length+i] for i in range(0, len(string), length))
                
        
while True:
        
        print(Menu)
       
        choice = input("\nEnter your choice (0/1/2/3/4/5/6/exit) : ")
        
        if choice == '0':
            count = '1'
        else:
            count = choice 
            
    
        if choice in options[0:-1]:
                try :
                    x = float(input("\nEnter the first number: "))

                    if choice != options[-2]: 
                        y = float(input("Enter the second number: "))

                except ValueError:
                    print("Please enter a valid number !!!")
                    print (separate)
                    continue

                calcul = Calculator(x,y)

                Display(calcul,count)       


        elif choice == options[-1]:
            print ("Bye Bye !!")
            break;

        else:
            print('\n choose a valid option !! \n')

        print (separate)
```
